```python
import pandas as pd
import numpy as np
import pickle
```

```python
data = pd.read_csv('weather_forecast_data.csv')
data 
```

```python
# check if there is any null value
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data['Rain'] = encoder.fit_transform(data['Rain'])
```

```python
data.head()
```

```python
x = data[['Temperature(celcius)','Relative_Humidity(percentage)','Wind_Speed (km/hr)','Cloud_Cover(percentage)','Pressure(milibars)']]   #independent feature
y = data['Rain']        # dependent feature
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=42) 
```

```python
from sklearn.preprocessing import StandardScaler 
```

```python
scalar = StandardScaler()
x_train = scalar.fit_transform(x_train)
x_test = scalar.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
```

```python
logistic.fit(x_train,y_train)
```

```python
y_pred = logistic.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score = accuracy_score(y_test,y_pred)
print('the score is ',score)
confusion_matrix = confusion_matrix(y_test,y_pred)
print('the confusion matrix in', confusion_matrix)
classification_report = classification_report(y_test,y_pred)
print('classification reports are',classification_report)
```

# Prediction for the new data

```python
temp = 25.4
relative_humdity = 75.46
wind_speed= 4.5
cloud_cover=20.2
pressue = 1013.2

output = logistic.predict(scalar.transform([[temp, relative_humdity, wind_speed, cloud_cover,pressue]]))
output = encoder.inverse_transform(output)
print('Temperature(celcius) = 25.4')
print('Relative_Humidity(percentage) = 75.46')
print('Wind_Speed (km/hr) = 4.5')
print('Cloud_Cover(percentage) = 20.2')
print('Pressure(milibars) = 1013.2')
print('Weather Prediction =',output[0]) 
```

# Saving model into pickle file

```python
pickle.dump(logistic,open('logistic_regression_model.pkl','wb'))
pickle.dump(scalar,open('scaler.pkl','wb'))
pickle.dump(encoder,open('label_encoder.pkl','wb'))
```
